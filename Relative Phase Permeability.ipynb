```python
import numpy as np
import pandas as pd
from plotly import graph_objects as go
import seaborn as sns
import plotly.express as px

from plotly.subplots import make_subplots
```

## Все скважины

```python
path=''
File_name = 'График ОФП Никита.xlsx'
OFPW = pd.read_excel(path + File_name, sheet_name='Вода-Нефть')
```

```python
sw = OFPW['sw']

krw = OFPW['krw']
krow = OFPW['krow']
pcow = OFPW['pcow']


fig = make_subplots(specs=[[{"secondary_y": True}]])


fig.add_trace(
    go.Scatter(
        x=sw,
        y=pcow, name="Капиллярное давление", line_color = 'green'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(
        x= sw,
        y=krw, name="ОФП воды", line_color = 'blue'),
    secondary_y=True,
)

fig.add_trace(
    go.Scatter(
        x=sw,
        y=krow, name="ОФП нефти", line_color = 'red'),
    secondary_y=True,
)


fig.update_layout(
    # title="Пористость, д.е.", 
    title_x = 0.5, titlefont=dict(family="arial",size=23,color="black") ,autosize=False,width=800, 
                  height=700,margin=dict(l=50,r=50,b=50,t=50,pad=4),paper_bgcolor="White", plot_bgcolor="White")

fig.update_xaxes(title_text="Водонасыщенность, д.е.", titlefont=dict(family="arial",size=23,color="black"), range=[0.2, 1.05],
                 showline=True, linewidth=2, linecolor='black', gridcolor='lightgray', tickfont=dict(family='Arial', color='black', size=18))

fig.update_yaxes(title_text="Давление, бар", titlefont=dict(family="arial",size=23,color="black"), secondary_y=False, range=[None, 4.9],
                 showline=True, linewidth=2, linecolor='black', gridcolor='lightgray', tickfont=dict(family='Arial', color='black', size=18))

fig.update_yaxes(title_text="Относительная проницаемость, д.е.", titlefont=dict(family="arial",size=23,color="black"), range=[None, 0.59], secondary_y=True, dtick=0.05,#  dtick=0.05 ШАГ
                 showline=True, linewidth=2, linecolor='black', tickfont=dict(family='Arial', color='black', size=18))


fig.update_layout(showlegend=True)

fig.update_layout( 
    legend=dict( 
        x=0.15,  # value must be between 0 to 1. 
        y=0.95,   # value must be between 0 to 1. 
        traceorder="normal", 
        font=dict( 
            family="arial", 
            size=20, 
            color="black"
        ), 
    ) 
) 

fig.show()

#fig.write_image(r"C:\Users\nik_c\OneDrive\Рабочий стол\Учеба\FDP\ОФП\OFP_oil.png")
```
