```python
import pandas as pd
from plotly import graph_objects as go
import numpy as np
```

Построение графиков Лоренца по скважинным данным

## R1

```python
Well = pd.read_excel('Statistics.xlsx', sheet_name='R4 (2)')
```

```python
Well['Мощность'] = Well['DEPTH']
for i in range(len(Well)):
    Well.loc[i, 'Мощность'] = Well.loc[i+1,'DEPTH'] - Well.loc[i,'DEPTH']
    if i == len(Well)-2 :
        Well.loc[len(Well)-1, 'Мощность'] = 0
        break
Well = Well.drop(len(Well)-1, axis=0)
Well['phi * h'] = Well['DEPTH']
for i in range(len(Well)):
    Well.loc[i, 'phi * h'] = Well.loc[i,'Por'] * Well.loc[i,'Мощность']

Well['k * h'] = Well['DEPTH']
for i in range(len(Well)):
    Well.loc[i, 'k * h'] = Well.loc[i,'PERMX'] * Well.loc[i,'Мощность']

Well['k / h'] = Well['DEPTH']
for i in range(len(Well)):
    Well.loc[i, 'k / h'] = Well.loc[i,'PERMX'] / Well.loc[i,'Por']

Well = Well.sort_values(by=['k / h'], ascending=False)

df = Well.copy()
# df = df.drop()
df = Well.reset_index().drop(['index'], axis=1)

df['phi * h Norm'] = df['DEPTH']
for i in range(len(df)):
    df.loc[i, 'phi * h Norm'] = sum(df.loc[0:i, 'phi * h'])/sum(df.loc[:, 'phi * h'])

df['k * h Norm'] = df['DEPTH']
for i in range(len(df)):
    df.loc[i, 'k * h Norm'] = sum(df.loc[0:i, 'k * h'])/sum(df.loc[:, 'k * h'])
df['phi * h Norm 2'] = df['DEPTH']
for i in range(len(df)):
    df.loc[i, 'phi * h Norm 2'] = (df.loc[i, 'phi * h Norm']-min(df.loc[:, 'phi * h Norm']))/(max(df.loc[:, 'phi * h Norm'])-min(df.loc[:, 'phi * h Norm']))

df['k * h Norm 2'] = df['DEPTH']
for i in range(len(df)):
    df.loc[i, 'k * h Norm 2'] = (df.loc[i, 'k * h Norm']-min(df.loc[:, 'k * h Norm']))/(max(df.loc[:, 'k * h Norm'])-min(df.loc[:, 'k * h Norm']))

s = 0
for j in range(len(df)):
    if j == 0:
        continue
    s += (df.loc[j, 'k * h Norm 2'] * (df.loc[j, 'phi * h Norm 2']-df.loc[j-1, 'phi * h Norm 2']))
s = (s-0.5)/0.5

x1= np.array([0,1])
y1= np.array([0,1])

fig = go.Figure()
fig.add_trace(go.Scatter(x=x1,y=y1, mode='lines', line_width = 4, line_color = 'red'))
fig.add_trace(go.Scatter(x=df.iloc[:,9],y=df.iloc[:,10], mode='markers', marker=dict(color='LightSkyBlue', size=3, 
                                                                                     line=dict(color='RGB(0,156,222)', width=3)), 
                                                                                     fill = 'tonexty', fillcolor=  '#FFBD73'))
fig.update_xaxes(title_text="Storativity", titlefont=dict(family="arial",size=22,color="black"))
fig.update_yaxes(title_text="Transmissivity", titlefont=dict(family="arial",size=22,color="black"))
fig.update_layout(autosize=False,width=750,height=700,margin=dict(l=50,r=50,b=50,t=50,pad=4),paper_bgcolor="White", plot_bgcolor = "White")
fig.update_xaxes(showline=True, linewidth=2, linecolor='black', gridcolor='lightgray',tickfont=dict(family='Arial', color='black', size=18))
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', gridcolor='lightgray',tickfont=dict(family='Arial', color='black', size=18))
fig.update_layout(title = 'Скважина R4', titlefont=dict(family='arial',size=24,color="black"), title_x = 0.5)
fig.update_yaxes(range = [0,1])
fig.update_xaxes(range = [0,1])
fig.update_layout(showlegend=False)
fig.add_annotation(x=0.86, y=0.05, text=f"Lc = {round(s, 2)}", font=dict(family='arial',size=40,color="black"), arrowwidth=4, arrowcolor='black',
                showarrow=True, xanchor='center',yanchor='top', ax = 0, ay = -17, arrowhead=0)
fig.show()
```
