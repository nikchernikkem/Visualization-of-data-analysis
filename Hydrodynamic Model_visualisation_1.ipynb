```python
df = pd.read_excel('рафики по лучшему варианту.xlsx')

# col = df.columns.to_list()
# for i in range(2,len(col)):
#     col[i]='Y'+str(i-1)
# df.columns=col

wells = ['X1','X2','X3','X4','X5','X6']
for j in wells:
    for i in range(2, len(df.columns), 2):


        fig = go.Figure()

        # for i in range(2,df.shape[1]-2): 
        #     if i not in index: 
        #         fig.add_trace(go.Scatter (x = df['Unnamed: 0'], y = df["Y"+str(i)]/1000, line = dict (color = 'gray'), opacity = 0.35, name = str(i+2)))

        # for i in range(2,df.shape[1]-2): 
        #     if i in index: # Выделение 10 кривых с наименьшей среднеквадратичной ошибкой
        #         

        fig.add_trace(go.Scatter( x = df.iloc[:,1], y = df.iloc[:,i+1][df.iloc[:,0]==j], mode = 'markers', opacity = 1, # Историческая кривая
                            line = dict (color = 'Black', width = 2, dash = 'dash'), name = 'History')) 
        
        fig.add_trace(go.Scatter (x = df.iloc[:,1], y = df.iloc[:,i][df.iloc[:,0]==j], opacity = 1, line = dict (width = 2), name = str(i)))
        
        fig.add_vline(x='2021-09-01', y1=1, xanchor="left", yanchor="bottom", line_color="rgb(0,152,216)", line_dash="dash")
        fig.update_layout(title = f'{df.columns[i]} по скважине {j}', titlefont = dict(size = 30,color = "black"), title_x = 0.5)
        fig.update_xaxes(title_text="Дата", titlefont=dict(family="arial",size=20,color="black"))
        if i == 10:
            fig.update_yaxes(title_text=f"{df.columns[i]}, бар", titlefont=dict(family="arial",size=20,color="black"))
        else:
            fig.update_yaxes(title_text=f"{df.columns[i]}, ст.м\u00b3/сут", titlefont=dict(family="arial",size=20,color="black"))    
        fig.update_layout(autosize=False, width=1200,height=800, margin=dict(l=50,r=50,b=50,t=50,pad=4), paper_bgcolor="White", plot_bgcolor="White")
        fig.update_xaxes(range=['2020-09-01', '2022-09-01'],showline=True, linewidth=2, linecolor='black', gridcolor='lightgray', tickfont=dict(family='Arial', color='black', size=17))
        fig.update_yaxes(showline=True, linewidth=2, linecolor='black', gridcolor='lightgray', tickfont=dict(family='Arial', color='black', size=17))

        fig.update_layout(showlegend=False)

        fig.show()

        # fig.write_image(f'output_test2\{df.columns[i][0:12]}_rate_{j}.png')
```
