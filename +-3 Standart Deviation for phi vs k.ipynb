Строим матрицу поворота:

$$
M(\theta) = 
\begin{pmatrix}
\cos \theta & \mp \sin \theta \\
\pm \sin \theta & \cos \theta
\end{pmatrix}
$$


$$
\begin{bmatrix}
x' \\
y'
\end{bmatrix}
=
\begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

```python
arr = np.array([df.iloc[:,2],df.iloc[:,5]]) # 5 - проницаемость иск, 2 - пористость игрек
arr = np.transpose(arr)
arr.shape
```

```python
ang = 26.13718685
rot = np.array([[np.cos(ang*3.14/180), -np.sin(ang*3.14/180)], 
                [np.sin(ang*3.14/180), np.cos(ang*3.14/180)]])
rot
```

```python
new_arr = arr.dot(rot)
new_arr = pd.DataFrame(new_arr)
new_arr
```

```python
STD_POR = np.std(new_arr.iloc[:,1]) #Density Ox
MEAN_POR = np.mean(new_arr.iloc[:,1])
MEAN_POR
```

```python
new_arr['-3std'],new_arr['+3std'] = new_arr[1], new_arr[1]
new_arr
```

```python
for i in range(len(new_arr['-3std'])):
    new_arr.iloc[i,2] = MEAN_POR-3*STD_POR
for i in range(len(new_arr['+3std'])):
    new_arr.iloc[i,3] = MEAN_POR+3*STD_POR
new_arr 
```
